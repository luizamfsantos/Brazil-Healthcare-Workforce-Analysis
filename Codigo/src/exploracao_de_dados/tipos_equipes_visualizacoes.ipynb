# import libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

# load data

```python
path = '../../../Dados/Processado/Equipes_Saude/totais/2007-2023_tipos_equipe.csv'
df = pd.read_csv(path)
```

```python
df.head()
```

```python
df.set_index('Tipo da Equipe', inplace=True)
```

```python
df.columns = pd.to_datetime(df.columns)
```

# visualizations

```python
# Plotando o grÃ¡fico
plt.figure(figsize=(18, 6))

# Loop para plotar os grÃ¡ficos
for equipe in df.index:
    plt.plot(df.columns, df.loc[equipe], label=equipe)

plt.xlabel('Ano')
plt.ylabel('NÃºmero de Equipes')
plt.title('NÃºmero de Equipes de SaÃºde por Tipo')
# plt.legend(loc = 'upper left', ncol =3)
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
